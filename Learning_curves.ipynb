```python
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np        
import time
import pandas as pd
import seaborn as sns
```

```python
df1 = pd.read_csv("experiments/iAC_lr_0.01.csv")
df2 = pd.read_csv("experiments/iAC_lr_0.05.csv")
df3 = pd.read_csv("experiments/iAC_lr_0.075.csv")
```

```python
R1 = sum([list(df1[i]/10) for i in df1.columns],[])
R2 = sum([list(df2[i]/10) for i in df2.columns],[])
R3 = sum([list(df3[i]/10) for i in df3.columns],[])
R = R1+R2+R3
Epoch1 = [i for i in range(201)]*20
Epoch = Epoch1+Epoch1+Epoch1
lr1 = [0.01]*len(R1)
lr2 = [0.05]*len(R2)
lr3 = [0.075]*len(R2)
lr = lr1+lr2+lr3
d = {'R':R,'Epoch':Epoch,'lr':lr} 
df = pd.DataFrame.from_dict(d)
```

```python
##скорость обучения
```

```python
plt.subplots(figsize=(12, 10))
sns.lineplot(data=df, x="Epoch", y="R",hue="lr",palette = 'vlag')
```
